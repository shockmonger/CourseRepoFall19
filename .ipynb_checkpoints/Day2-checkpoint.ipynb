# Creating a dataset with features

```python
import pandas as pd
# importing pandas for being able to load data

from pathlib import Path
# to easily write a function to work on all the files in a folder

kick_signals = [    
    librosa.load(p)[0] for p in Path().glob('Data/drum_samples/*/kick*.mp3')
]
snare_signals = [
    librosa.load(p)[0] for p in Path().glob('Data/drum_samples/*/snare*.mp3')
]
```

```python
def extract_features(signal):
#This is the function definition
    return [
        np.mean(librosa.feature.zero_crossing_rate(signal)),
        np.mean(librosa.feature.spectral_centroid(signal)),
        np.mean(librosa.feature.spectral_contrast(signal)),
        np.mean(librosa.feature.chroma_stft(signal)),
        np.mean(librosa.feature.rmse(signal)),
        np.mean(librosa.feature.mfcc(signal)),
        np.mean(librosa.feature.spectral_flatness(signal)),
        np.mean(librosa.feature.tempogram(signal))
    ]
```

```python
kick_features = np.array([extract_features(x) for x in kick_signals])
snare_features = np.array([extract_features(x) for x in snare_signals])
print(kick_features.shape, snare_features.shape)
kicklabels = np.zeros(72)
snarelabels = np.ones(42)
labels = np.concatenate((kicklabels,snarelabels))
print(labels.shape)

features = np.concatenate((kick_features,snare_features))
featureset = pd.DataFrame(features)
featureset.columns = ['ZC','SpecCen','SpecCon','Chroma','RMSE','MFCC','SpecFlat','Tempogram']
featureset
```

```python
scaled_featureset = sklearn.preprocessing.scale(featureset)
scaled_featureset = pd.DataFrame(scaled_featureset)
scaled_featureset.columns = ['ZC','SpecCen','SpecCon','Chroma','RMSE','MFCC','SpecFlat','Tempogram']
scaled_featureset['Labels'] = labels
scaled_featureset
```

# Principle Component Analysis

Focus on describing the technique and the library


Principal component analysis (PCA) is a procedure that transforms a number of (possibly) correlated variables into a (smaller) number of uncorrelated variables called principal components. 
The first principal component accounts for as much of the variability in the data as possible, and each succeeding component accounts for as much of the remaining variability as possible.

# K Nearest Neighbors

### Load the list of sound files

Make sure you specify the path in the say that applies to the directory structure applicable for your operating system. 

```python
# Our data is stored at the variable> scaled_featureset
pd.DataFrame(scaled_featureset)

#we need data without labels, which is the last column, we define this in the next var>
df_feat = pd.DataFrame(scaled_featureset.iloc[:,-1])
```

```python
from sklearn.model_selection import train_test_split
# Set the X and ys
X = df_feat
y = labels
# Use the train_test_split() method to split the data into respective sets
# test_size -> argument refers to the size of the test subset
# random_state -> argument ensures guarantee that the output of Run 
# 1 will be equal to the output of Run 2, i.e. your split will be always the sameX_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Import module for KNN
from sklearn.neighbors import KNeighborsClassifier

# Create KNN instance
# n_neighbors -> argument identifies the amount of neighbors used to ID classification

knn = KNeighborsClassifier(n_neighbors=1)
# Fit the model (this step is also called training)

knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Print out classification report and confusion matrix
print(classification_report(y_test, pred))
```

```python
# Print out confusion matrix
cmat = confusion_matrix(y_test, pred)#print(cmat)
print('TP - True Negative {}'.format(cmat[0,0]))
print('FP - False Positive {}'.format(cmat[0,1]))
print('FN - False Negative {}'.format(cmat[1,0]))
print('TP - True Positive {}'.format(cmat[1,1]))
print('Accuracy Rate: {}'.format(np.divide(np.sum([cmat[0,0],cmat[1,1]]),np.sum(cmat))))
print('Misclassification Rate: {}'.format(np.divide(np.sum([cmat[0,1],cmat[1,0]]),np.sum(cmat))))
```

```python
# Generate function to add error rates of KNN with various k-values
# error_rate -> empty list to gather error rates at various k-values
# for loop -> loops through k values 1 to 39
# knn -> creates instance of KNeighborsClassifier with various k
# knn.fit -> trains the model
# pred_i -> conducts predictions from model on test subset
# error_rate.append -> adds error rate of model with various k-value, using the average where prediction not
# equal to the test valueserror_rate = []for i in range(1,40):

error_rate = []
# we will assign i nearest neighbors a value

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# Configure and plot error rate over k valuesplt.figure(figsize=(10,4))
plt.plot(range(1,40), error_rate, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K-Values')
plt.xlabel('K-Values')
plt.ylabel('Error Rate')
```
